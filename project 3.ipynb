```python
df = pd.read_csv('Suicide Rate.csv',on_bad_lines='skip',delimiter=',')
df.dropna(inplace=True)
print(df.to_string())
```

```python
import numpy as np
from scipy import stats
from statsmodels.stats import weightstats as stests
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
import seaborn as sns
import scipy
```

1) Null Hypothesis = There is no relation between GDP and suicide rates

2)Alternative Hypothesis = There is a negative or opposite relation between suicide rates and GDP


```python
#implementing the correlation test and generating the p-value and correlation coefficient
import scipy.stats as stats
suicide_rates = df['Suicide rate']
gdp = df['GDP per capita']
correlation_coefficient, p_value = stats.pearsonr(suicide_rates, gdp)
print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)
```

P-value is smaller than 0.05. Hence, we reject the null hypothesis.

Correlation coefficient is -0 which means that there is a weak negative correlation, meaning as gdp increases the suicide rates decreases ,and vice versa, but the relationship is not very strong.

```python
#visiualize the correlation between the GDP and suicide rates
variable1 = df['Suicide rate']
variable2 = df['GDP per capita']

# Create a scatter plot
plt.scatter(variable1, variable2)
plt.xlabel('Suicide rates')
plt.ylabel('GDP')
plt.title('Correlation Scatter Plot')

# Add a regression line
sns.regplot(x=variable1, y=variable2, color='lightblue')

# Show the plot
plt.show()
```

```python

# Create a bar chart that visualizes the relationship between countries and their suicide rates

# Sort the DataFrame by 'GDP' in descending order
df_sorted = df.sort_values('GDP per capita', ascending=False)

# Create the bar plot
plt.figure(figsize=(21, 6))
plt.bar(df_sorted['Country'], df_sorted['Suicide rate'], color = 'lightblue')

# Customize the plot
plt.xlabel('Countries sorted by GDP ascendingly')
plt.ylabel('Suicide Rates')
plt.title('Suicide Rates by Country')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```
